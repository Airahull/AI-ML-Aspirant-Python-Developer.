```python
#for selecting rows by name 
#loc meaning is label based indexing 
df1['name'] = df1['name'].str.strip() # remove the void spaces from it using strip()
df1.loc[df1['name'] == 'rahul']#then using loc function 
```

```python
df1.loc[(df1['name'] == 'rahul') & (df1['monthly_salary'] > 150000)]
 # this is basically for printing rahul name persons in list who's salary is higher then 150000 and we using & operator for condition 
```

```python
#selecting value via index value 
df1.iloc[1]
```

```python
#we can also use slicing on it 
df1.iloc[0:2]# iloc does'nt takes 2 row  
```

```python
df1.loc[0:2]# loc method takes 2 row as well 
```

```python
# filter dataframe 
df1['age '] > 24 # it gives boolean value in output 
```

```python
#if we want actual row which satisfied condition 
df1[df1['age '] >24]
```

```python
# we can also store this filter dateframe into new dataframe 
filtered_by_age_df = df1[df1['age '] >24]
filtered_by_age_df
```

```python
# we can also filter value on the basic of 2 or more condition 
df1[(df1['age '] > 24)&(df1['monthly_salary'] >48000)]
```

```python
# if we need to filter value with changing the size of dataframe 
df1.where(df1['age '] > 26) # we use df.where() function
```

```python
# we also replace nan with anything 
df1.where(df1['age '] >=26 , other = 'Not Eligible')
```
