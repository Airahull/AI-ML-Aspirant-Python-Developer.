```python
#slicing in 1d array a[start:stop:steps]
aa=np.array([1,2,3,4,5])
print(aa[0:3]) # first three element 
print(aa[-3:]) # last three element 
print(aa[-4:-1]) # middle three element 
print(aa[0::2]) # whole array with steps 
```

```python
#array reshaping and flattening 
#1. array reshaping accoring to user
arrr =np.array([1,2,3,4,5,5,6,7])
print(arrr)
print("reshaped array")
reshaped_arr = arrr.reshape([4,2])#reshape in 4 rows and 2 colowmn 
print(reshaped_arr)
reshaped_arr = arrr.reshape([2,4])#reshape in 2 rows and 4 colowmn 
print(reshaped_arr)
```

```python
#flattening a array 
print("before flatten :")
print(reshaped_arr)
print("after flatten :")
print(reshaped_arr.flatten())
```

```python
#array stacking 
#1.vertical stacking 
a1=np.array([1,2,3,4])
a2=np.array([5,6,7,8])
print(np.vstack((a1,a2)))# vertical stacking row-wise // and converted in 2d array 
#2.horizontal stacking 
print("\n")
print(np.hstack((a1,a2))) # horizontal stacking colowmn-wise // remain same 1d array  
```

```python
#array splitting 
#1.vertical splitting 
c=np.array([[1,2,3],[4,5,6]]) #split the row  
print(c)
print(np.vsplit(c,2))
```

```python
#2.horizontal spilting 
hsplit = np.hsplit(c,3) #split the colwmn
for ar in hsplit:
    print(ar)
    
```
